## This notebook aims to create the final df from needed to test the AI models


```python
import pandas as pd;
```

```python
df = pd.read_csv('clean_london_all_2017_trip_data.csv')
df
```

```python
df['Start Date'] = pd.to_datetime(df['Start Date'], dayfirst=True, format="mixed", errors='raise') #'%d/%m/%Y %H:%M',
df.sort_values(by=['Start Date'], inplace=True)
df
```

```python
df['StartDate Rounded'] = df['Start Date'].dt.round("h")

min_station_id = df['StartStation Id'].min()
total_stations = df['StartStation Id'].max()

fulldf = pd.DataFrame()
dfs = []
for id in range(min_station_id, min_station_id + total_stations):
    hourly_counts = df.loc[df['StartStation Id'] == id].groupby(['StartDate Rounded', 'StartStation Id']).size().reset_index(name='trips')
    # print(id)
    # print(hourly_counts.shape)
    # print(hourly_counts)
    dfs.append(hourly_counts)

fulldf = pd.concat(dfs, ignore_index=True)
```

```python
fulldf['StartDate Rounded'] = pd.to_datetime(fulldf['StartDate Rounded'], dayfirst=True,format="mixed", errors='raise') #'%Y/%m/%d %H:%M:%S'
fulldf['StartHour Rounded'] = fulldf['StartDate Rounded'].dt.hour
```

```python
fulldf["StartDate Rounded"] = pd.to_datetime(fulldf["StartDate Rounded"],dayfirst=True, format="mixed", errors = 'raise')
```

```python
#fulldf['StartDate Rounded'] = pd.to_datetime(fulldf['StartDate Rounded'], format='%Y-%m-%d %H:%M:%S', errors='coerce') #'%Y/%m/%d %H:%M:%S'
fulldf = fulldf.sort_values(by=['StartDate Rounded'])
# weather_df['time'] = pd.to_datetime(weather_df['time'], format='%Y-%m-%dT%H:%M', errors='coerce')
# weather_df.sort_values(by=['time'], inplace=True)
#fulldf.reset_index(drop=True, inplace=True)
#test["StartHour Rounded"] == 1
#test.to_csv("test.csv", index=False)


fulldf['New_DateTime'] = fulldf['StartDate Rounded'].dt.strftime('%Y-%m-%d') + ' ' + fulldf['StartHour Rounded'].astype(str)
```

```python
fulldf.drop(columns=["StartDate Rounded", "StartHour Rounded"], inplace=True)
fulldf.rename(columns={"New_DateTime":"StartDate"}, inplace=True)
fulldf["StartDate"] = pd.to_datetime(fulldf['StartDate'], format="%Y-%m-%d %H")
```

```python
weather_df = pd.read_csv("london_2017_weather_data_hourly&daily.csv")

weather_df.rename(columns={"temperature_2m (Â°C)":"real_temp",
                          "apparent_temperature (Â°C)":"feel_temp",
                           "relative_humidity_2m (%)": "humidity",
                           "precipitation (mm)":"precipitation",
                           "rain (mm)":"rain",
                           "snowfall (cm)":"snowfall",
                           "weather_code (wmo code)":"weather_code",
                           "cloud_cover (%)":"cloud_cover",
                           "wind_speed_100m (km/h)":"wind_speed"
                          }, inplace=True)

weather_df.head()
```

```python
weather_df['time'] = pd.to_datetime(weather_df['time'], format='%Y-%m-%dT%H:%M', errors='coerce')
weather_df.sort_values(by=['time'], inplace=True)
weather_df
```

```python
merged_df = pd.merge_asof(fulldf, weather_df, direction='nearest', left_on='StartDate', right_on='time')
merged_df.drop(['time'], axis=1, inplace=True)
merged_df
```
